```python
from tkinter import *
import math

expression = ""
ans = ""
history = []

def press(num):
    global expression
    if equation.get() == " error ":
        clear()
    cursor_position = expression_field.index(INSERT)
    if num in ['sin', 'cos', 'tan', 'log', 'asin', 'acos', 'atan', 'sqrt']:
        num += '('
    expression_field.insert(cursor_position, str(num))
    expression = expression_field.get()
    equation.set(expression)

def equalpress():    
    global ans
    try:
        global expression   
        total = str(eval(expression))
        equation.set(total)
        ans = total
        history.append(expression + " = " + total)
        expression = ""
    except:
        equation.set(" error ")
        expression = ""
        blink_error()

def clear(): 
    global expression 
    expression = "" 
    equation.set("") 

def blink_error():
    current_color = expression_field.cget("background")
    new_color = "red" if current_color == "white" else "white"
    expression_field.config(background=new_color)
    if equation.get() == " error ":
        expression_field.after(500, blink_error)

def show_history():
    history_window = Toplevel(gui)
    history_window.title("History")
    history_window.geometry("400x400")
    history_text = Text(history_window, font=('arial', 12, 'bold'))
    history_text.pack(expand=True, fill=BOTH)
    for item in history:
        history_text.insert(END, item + "\n")

if __name__ == "__main__": 
    gui = Tk()
    gui.configure(background="sky blue") 
    gui.title("Scientific Calculator") 
    gui.geometry("800x800")

    equation = StringVar() 

    expression_field = Entry(gui, textvariable=equation, font=('arial', 20, 'bold')) 
    expression_field.grid(row=0, column=0, columnspan=6, ipadx=70, sticky="nsew")

    button_params = {'fg': 'black', 'bg': 'light green', 'height': 2, 'width': 10, 'font': ('arial', 18, 'bold')}
    
    buttons = [
        ('1', 1, 2, 0), ('2', 1, 2, 1), ('3', 1, 2, 2), ('+', 1, 2, 3),
        ('4', 1, 3, 0), ('5', 1, 3, 1), ('6', 1, 3, 2), ('-', 1, 3, 3),
        ('7', 1, 4, 0), ('8', 1, 4, 1), ('9', 1, 4, 2), ('*', 1, 4, 3),
        ('0', 1, 5, 0), ('.', 1, 5, 1), ('(', 1, 5, 2), (')', 1, 5, 3),
        ('%', 1, 6, 0), ('**', 1, 6, 1), ('//', 1, 6, 2), ('/', 1, 6, 3),
        ('sin', 1, 7, 0), ('cos', 1, 7, 1), ('tan', 1, 7, 2), ('log', 1, 7, 3),
        ('asin', 1, 8, 0), ('acos', 1, 8, 1), ('atan', 1, 8, 2), ('sqrt', 1, 8, 3),
        ('Ans', 1, 9, 0), ('History', 1, 9, 1), ('=', 1, 9, 3), ('Clear', 1, 9, 2)
    ] 

    for (text, rowspan, row, col) in buttons:
        Button(gui, text=text, command=lambda t=text: press(t) if t not in ['=', 'Clear', 'Ans', 'History'] else equalpress() if t == '=' else clear() if t == 'Clear' else press(ans) if t == 'Ans' else show_history(), **button_params).grid(row=row, column=col, rowspan=rowspan, sticky="nsew")

    for i in range(10):
        gui.grid_rowconfigure(i, weight=1)
    for i in range(6):
        gui.grid_columnconfigure(i, weight=1)

    gui.mainloop()
```
